```python
import time
import os
import numpy as np
import pandas as pd
import cv2
import json
import shutil
```

```python
IN_DIR = "/datadrive/workspace/others/202206.tanachou/data/train"
OUT_DIR_1 = "/datadrive/workspace/others/202206.tanachou/data/train_category"
OUT_DIR_2 = "/datadrive/workspace/others/202206.tanachou/data/train_color"
```

```python
json_open = open('../data/train_meta.json', 'r')
df_train = pd.DataFrame(json.load(json_open)).T
```

```python
df_train
```

```python
for dir_name in os.listdir(IN_DIR):
    category = df_train.loc[dir_name].category
    color = df_train.loc[dir_name].color
    
    in_dir = "{}/{}".format(IN_DIR, dir_name)
    out_dir = "{}/{}".format(OUT_DIR_1, category)
    if not os.path.exists(out_dir):
        os.makedirs(out_dir)
    for img_name in os.listdir(in_dir):
        img_path = "{}/{}".format(in_dir, img_name)
        new_img_name = "{}_{}".format(dir_name, img_name)
        new_img_path = "{}/{}".format(out_dir, new_img_name)
        
        shutil.copy(img_path, new_img_path)
```
