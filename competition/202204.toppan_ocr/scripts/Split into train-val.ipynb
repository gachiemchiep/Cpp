```python
import glob 

toppan_ocr_dir = "/datadrive/workspace/others/202203.toppan_ocr/dataset/train_r90"
txt_paths = glob.glob("{}/*/*.txt".format(toppan_ocr_dir))
toppan_texts = []
for txt_path in txt_paths:
    with open(txt_path, "r") as fid:
        text = fid.readline().rstrip()
        toppan_texts.append(text)
        
toppan_all_text = "".join(toppan_texts)
toppan_all_chars = sorted(set(toppan_all_text))
```

```python
char_occ_dict = {}
for char in toppan_all_chars:
    occ = toppan_all_text.count(char)
    char_occ_dict[char] = occ
    
    print("{} : {}".format(char, occ))
```
